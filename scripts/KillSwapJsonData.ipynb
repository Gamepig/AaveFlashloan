```python
import json
NewData = []
DataCount = 0
with open("SwapLog.json", "r") as file:
    data = json.load(file)

for swapinfo in data:
    profit = float(swapinfo['Profit'])
    if(profit>50):
        NewData.append(swapinfo)
        DataCount += 1
# with open("SwapLog.json", "w") as file:
#     json.dump(NewData, file, indent=4)
print(NewData)
print(DataCount)
```
